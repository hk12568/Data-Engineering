```python
import requests
from scrapy import Selector
import pandas as pd
```

# Step 1: Fetch Data from the Website


```python
url="https://www.bikewale.com/royalenfield-bikes/"
```

```python
headers=({'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:135.0) Gecko/20100101 Firefox/135.0'})
```

```python
response = requests.get(url, headers = headers)
```

```python
response
```

# Step 2: Parse HTML using Scrapy Selector


```python
bike=Selector(text=response.text)
```

# Ensure all lists are of the same length


```python
min_length = min(len(name), len(price), len(rating), len(star), len(showroom), len(kg), len(cc), len(kmpl), len(bhp))
```

```python
min_length
```

```python
names = name[:min_length]
prices = price[:min_length]
ratings = rating[:min_length]
stars = star[:min_length]
showrooms = showroom[:min_length]
kgs = kg[:min_length]
ccs = cc[:min_length]
kmpls = kmpl[:min_length]
bhps = bhp[:min_length]

```

# Step 4: Store Data in a Pandas DataFrame


```python
data = pd.DataFrame({
    "Name": names,
    "Price": prices,
    "Rating": ratings,
    "Star": stars,
    "Showroom": showrooms,
    "KG ": kgs,
    "CC": ccs,
    "Mileage": kmpls,
    "BHP": bhps
})
```

```python
data
```

```python
data.to_csv("hari_enfield_bikes_unclean.csv",index=False)
```

# Step 5: Clean and Save the Data


```python
def convert_price(price):
    try:
        return float(price.replace("₹", "").replace(",", ""))
    except:
        return None

data["Price"] = data["Price"].apply(convert_price)

```

```python
data
```

```python
def convert_ratings(Rating):
    try:
        return float(Rating.replace("Ratings", "").replace(",", ""))
    except:
        return None

data["Rating"] = data["Rating"].apply(convert_ratings)
```

```python
data
```

```python
def convert_kg(kg):
    try:
        return float (kg.replace("kg", ""))
    except:
        return None

data['KG '] = data['KG '].apply(convert_kg)
```

```python
data
```

```python
def convert_cc(CC):
    try:
        return float(CC.replace("cc", "").replace(",", ""))
    except:
        return None

data["CC"] = data["CC"].apply(convert_cc)
```

```python
data
```

```python
def convert_kmpl(Mileage):
    try:
        return float(Mileage.replace("kmpl", "").replace(",", ""))
    except:
        return None

data["Mileage"] = data["Mileage"].apply(convert_kmpl)
```

```python
data
```

```python
def convert_bhp(BHP):
    try:
        return float(BHP.replace("bhp", "").replace(",", ""))
    except:
        return None

data["BHP"] = data["BHP"].apply(convert_bhp)
```

```python
data
```

```python
data.to_csv("hari_enfield_bikes.csv", index=False)
```
